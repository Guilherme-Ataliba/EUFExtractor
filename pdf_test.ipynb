```python
import fitz  # PyMuPDF
import re

def extract_text_from_page(pdf_path, page_num):
    """Extract text from a specific page using PyMuPDF."""
    doc = fitz.open(pdf_path)
    page = doc[page_num]  # Select page (0-based index)
    text = page.get_text("text")  # Extract text
    return text

def search_pattern_in_text(text, pattern):
    """Find all occurrences of a regex pattern in the extracted text."""
    matches = re.findall(pattern, text)
    return matches

# Example usage
pdf_path = "euf-2023-2.pdf"  # Replace with your PDF file
page_number = 1  # Change as needed
pattern = r"A(.*\n).*"  # Example: Extract codes like "ABC123"

extracted_text = extract_text_from_page(pdf_path, page_number)
matches = search_pattern_in_text(extracted_text, pattern)

print("Extracted Text:\n", extracted_text)
print("\nMatches Found:", matches)
```

```python
import fitz  # PyMuPDF
import re

def extract_alternatives(pdf_path, page_num):
    """Extract alternatives (A, B, C, D, E) from a given page."""
    doc = fitz.open(pdf_path)
    page = doc[page_num]
    text = page.get_text("text")  # Extract text
    
    # Regex pattern to match alternatives like "A L = ..." or "B ω0 = ..."
    pattern = r"\n([A-E])\s(.+?)(?=\n[A-E]|\nQuestão|\Z)"
    
    matches = re.findall(pattern, text, re.DOTALL)
    
    return {letter: content.strip() for letter, content in matches}

# Example usage
pdf_path = "euf-2023-2.pdf"
page_number = 0  # Change as needed

alternatives = extract_alternatives(pdf_path, page_number)
for key, value in alternatives.items():
    print(f"{key}: {value}\n")
```
