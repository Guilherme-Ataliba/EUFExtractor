```python
def clean_input(string):
    return string.replace(r"\begin{aligned}", "").replace(r"\end{aligned}", "").replace("&", "")
```

```python
def shuffle_and_save(inputs):
    alternatives = ["A", "B", "C", "D", "E"]

    new_inputs = {}
    correct_pos = {}

    for key, value in inputs.items():
        arr = [0, 1, 2, 3, 4]
        combined = list(zip(arr, value))
        np.random.shuffle(combined)
        arr, arr2 = zip(*combined)

        arr = list(arr)
        arr2 = list(arr2)

        correct_pos[key] = alternatives[arr.index(0)]
        new_inputs[key] = arr2

    return new_inputs, correct_pos
```

```python
filt_inputs = {}

final_print_table = ""
answers_dict = {}

counter = 0
for key, value in clean_inputs.items():
    filt_inputs[key] = value
    counter += 1

    if counter == 2: 
        new_inputs, correct_pos = shuffle_and_save(filt_inputs)
        answers_dict.update(correct_pos)

        table_string = pd.DataFrame(new_inputs)
        table_string.index=["A", "B", "C", "D", "E"]
        table_string = table_string.to_latex().replace(r"\begin{tabular}{lll}"+"\n", "").replace(r"\toprule"+"\n", "").replace(r"\midrule"+"\n", "").replace(r"\bottomrule"+"\n", "").replace(r"\end{tabular}"+"\n", "")
        final_print_table += table_string
        final_print_table += "\n" + r"\midrule" + "\n"
        
        counter = 0
        filt_inputs = {}

answers_df = pd.DataFrame([answers_dict]).T
answers_df.columns = ["Reposta"]
```
